## Get data

We want all data in a nicely formatted: title, real_or_fake format. Due to the limitation that our data comes from multiple data sources and come with multiple formats and columns, we need to try our best to integrate all data sources.

```python
import pandas as pd

# helper function
def prep_df(df,title_col,real_or_fake):
    df['real_or_fake'] = real_or_fake
    df['title'] = df[title_col].apply(lambda x: x.lower())
    df = df[['title', 'real_or_fake']]
    return df

# abc news
abc_news = pd.read_csv('data/abcnews-date-text.csv')
abc_news = prep_df(abc_news, 'headline_text', 1)
abc_news.head()
```

```python
gossip_fake = pd.read_csv('data/gossipcop_fake.csv', delimiter=',')
gossip_fake = prep_df(gossip_fake, 'title', 0)
gossip_fake.head()
```

```python
gossip_real = pd.read_csv('data/gossipcop_real.csv', delimiter=',')
gossip_real = prep_df(gossip_real, 'title', 1)
gossip_real.head()
```

```python
politifact_fake = pd.read_csv('data/politifact_fake.csv', delimiter=',')
politifact_fake = prep_df(politifact_fake, 'title', 0)
politifact_fake.head()
```

```python
politifact_real = pd.read_csv('data/politifact_real.csv', delimiter=',')
politifact_real = prep_df(politifact_real, 'title', 1)
politifact_real.head()
```
